```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from keras.datasets import mnist
from tensorflow.math import confusion_matrix
```

```python
(X_train,Y_train),(X_test,Y_test) = mnist.load_data()
```

```python
type(X_train)
```

```python
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
print(X_train[10])
```

```python
print(X_train[10].shape)
```

```python
plt.imshow(X_train[2001])
plt.show()
print(Y_train[2001])
```

```python
print(np.unique(Y_train))
print(np.unique(Y_test))
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
print(X_train[10])
```

```python
model = keras.Sequential([keras.layers.Flatten(input_shape = (28,28)), keras.layers.Dense(50,activation ='relu'), keras.layers.Dense(50,activation ='relu'), keras.layers.Dense(10,activation ='sigmoid')])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs = 10)
```

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(accuracy)
```

```python
plt.imshow(X_test[0])
plt.show()
print(Y_test[0])
```

```python
Y_pred = model.predict(X_test)
```

```python
print(Y_pred.shape)
```

```python
print(Y_pred[0])
```

```python
label_for_first_test_image = np.argmax(Y_pred[0])
print(label_for_first_test_image)
```

```python
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

```python
print(Y_pred)
```

```python
conf_mat = confusion_matrix(Y_test, Y_pred_labels)
```

```python
print(conf_mat)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(conf_mat, annot = True, fmt = 'd', cmap = 'Blues')
plt.ylabel('True Labels')
plt.xlabel('Predicted Labels')
```
